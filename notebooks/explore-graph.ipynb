```python
import os

from arango import ArangoClient
from dotenv import load_dotenv
```

```python
load_dotenv("../docker/dev/.env")

client = ArangoClient(hosts=f"http://localhost:{os.environ.get('ARANGO_PORT')}")

db = client.db(
    os.environ.get("ARANGO_DB"),
    username=os.environ.get("ARANGO_USER"),
    password=os.environ.get("ARANGO_ROOT_PASSWORD"),
)
```
